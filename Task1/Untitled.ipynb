```python
data = {"Отделение": [26, 26, 26],
        "Дата": ["01 Февраль 2019", "01 Февраль 2019", "01 Февраль 2019"],
        "Номер клиента": [1,2,3],
        "Время выдачи":["08:00:00", "08:00:00", "08:00:00"],
        "Время вызова":["08:01:49", "08:00:49", "08:01:10"],
        "Время завершения":["08:41:46", "08:02:23", "08:22:03"],
        "Длительность ожидания":["00:01:49", "00:00:49", "00:01:10"],
        "Длительность обслуживания":["00:39:57", "00:02:34", "00:20:53"],
        "Длительность нахождения в очереди":["00:41:46", "00:02:23", "00:22:03"],
        "Услуга": ["Подача документов физического лица", "Подача документов юридического лица", "Получение документов юридического лица"],
        "Оператор": ["Tula_26_1","Tula_26_10","Tula_26_5"],
        "Точка обслуживания": ["Tula_26_1","Tula_26_10","Tula_26_5"],
        "Последнее событие": ["Завершен","Не подтвержден", "Удален"],
        "Категория": [1, 2, 5],
        "Категория каждый день": [1, 2, 5]}
```

```python
df = pd.DataFrame.from_dict(data)
```

```python
df
```

```python
def ChangeDateFormat(name):
    
    months = ("Январь","Февраль","Март","Апрель","Май","Июнь","Июль","Август","Сентябрь","Октябрь","Ноябрь","Декабрь")
    day, mon, year = name.split(" ")
    if mon not in months:
        print("mon = ", mon)
    else:
        mon = months.index(mon)+1
    
    return f"{day}-0{mon}-{year}" if int(mon) < 10 else f"{day}-{mon}-{year}"
```

```python
df["Дата"] = df["Дата"].apply(ChangeDateFormat)
```

```python
df["Дата"] = pd.to_datetime(df["Дата"])
```

```python
df
```

```python
def TimeToSeconds(time):
    hours, mins, secs = map(lambda x: int(x), time.split(':'))
    return hours*3600+mins*60+secs
```

```python
def TimeToCategory(time):
    hours = int(time.split(':')[0]) - 8 + 1
    return hours
```

```python
def ServiceToCategory(title):
    services = ("Подача документов физического лица", "Подача документов юридического лица", "Получение документов юридического лица")
    
    if title not in services:
        print("service = ", title)
    else:
        title = services.index(title)
    
    return title
```

```python
df["Услуга"] = df["Услуга"].apply(ServiceToCategory)
```

```python
df
```

```python
def OperatorToCategory(oper):
    operators = ("Tula_26_1", "Tula_26_10", "Tula_26_5")
    
    if oper not in operators:
        print("operator = ", oper)
    else:
        oper = operators.index(oper)
        
    return oper
```

```python
df["Оператор"] = df["Оператор"].apply(OperatorToCategory)
```

```python
df["Точка обслуживания"] = df["Точка обслуживания"].apply(OperatorToCategory)
```

```python
def EventToCategory(event):
    events = ("Завершен","Не подтвержден","Удален")
    
    if event not in events:
        print("event = ", event)
    else:
        event = events.index(event)
        
    return event
```

```python
df["Последнее событие"] = df["Последнее событие"].apply(EventToCategory)
```
